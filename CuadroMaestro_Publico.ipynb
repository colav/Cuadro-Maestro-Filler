```python
#Libraries:
import requests
import gspread
from google.auth import default
import pandas as pd
import time
import gdown
import numpy
import random
import os

# Authenticate and connect to Google Sheets
from google.colab import auth
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload
from bs4 import BeautifulSoup
from collections import defaultdict
from requests.exceptions import RequestException
from openpyxl import load_workbook

# Create the Drive service
drive_service = build("drive", "v3")

#Regular expressions
import re

#Selenium Library
import google_colab_selenium as gs

#Libraries for web-scraping

from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import google_colab_selenium as gs
from selenium.common.exceptions import TimeoutException
from IPython.display import Image, display

#Main routine
def cuadro_maestro():
  download_files()

  # Extract group codes from the Cuadro Maestro
  codes = cdrom_df.iloc[:,2]
  codes = codes.astype(str).str.findall(r'COL\d+')
  codes = codes.explode().dropna()
  format_cdrom()
  print("Códigos:")
  print(codes)

  for i, code in enumerate(codes):
    print("Código de grupo: " + str(code))
    boletin_row = boletin_df.loc[boletin_df[' CÓDIGO GRUPO DE INVESTIGACIÓN'] == str(code)]
    url = str(boletin_row.iloc[0]['Enlace GrupLAC']).strip()
    name = str(boletin_row.iloc[0]['GRUPOS DE INVESTIGACIÓN'])
    soup = connect_gruplac(url)
    indexed_df, total_row = search_indexed_journals(soup, code)
    #Write collected info into dataframe
    cdrom_df.loc[i, 'NOMBRE DEL GRUPO DE INVESTIGACIÓN'] = name
    cdrom_df.loc[i, 'NÚMERO DE ARTÍCULOS EN REVISTAS INDEXADAS NACIONALES /Publindex'] = int (total_row.iloc[0]['Publindex'])
    cdrom_df.loc[i, 'NÚMERO DE ARTÍCULOS EN REVISTAS INDEXADAS INTERNACIONALES/Scopus /WoS'] = int(total_row.iloc[0]['Internacional'])
    cdrom_df.loc[i, 'NÚMERO DE LIBROS O CAPÍTULOS DE LIBROS DE INVESTIGACIÓN.'] = int(count_basic_prod(soup, target_header="Libros publicados") +
                                                                                      count_basic_prod(soup, target_header="Capítulos de libro publicados"))
    cdrom_df.loc[i, 'NÚMERO DE PATENTES, DISEÑOS INDUSTRIALES O REGISTROS DE SOFTWARE.'] = int(count_basic_prod(soup, target_header="Diseños industriales") +
                                                                                               count_basic_prod(soup, target_header="Softwares"))
    art_products = count_art_products(soup)
    cdrom_df.loc[i, 'NÚMERO DE PRODUCTOS DE CREACIÓN'] = int(art_products[0] + art_products[1] + art_products[2])
    thesis = count_thesis(soup)
    cdrom_df.loc[i, 'NÚMERO DE ESTUDIANTES VINCULADOS A TRABAJOS DE GRADO DEL PROGRAMA '] = int(thesis[0] + thesis[1] + thesis[2])

  display(cdrom_df)
  write_row_by_row(cdrom_df, "1Oc4ybTMWO-Z5YXMe3VFHtW3nknqy81Na", len(codes))


def format_cdrom():
  cols_to_int = [
    'NÚMERO DE ARTÍCULOS EN REVISTAS INDEXADAS NACIONALES /Publindex',
    'NÚMERO DE ARTÍCULOS EN REVISTAS INDEXADAS INTERNACIONALES/Scopus /WoS',
    'NÚMERO DE LIBROS O CAPÍTULOS DE LIBROS DE INVESTIGACIÓN.',
    'NÚMERO DE PATENTES, DISEÑOS INDUSTRIALES O REGISTROS DE SOFTWARE.',
    'NÚMERO DE PRODUCTOS DE CREACIÓN',
    'NÚMERO DE ESTUDIANTES VINCULADOS A TRABAJOS DE GRADO DEL PROGRAMA '
  ]

  cdrom_df[cols_to_int] = cdrom_df[cols_to_int].fillna(0).astype("int64")

#Safe download from Google Drive, managing error messages
def safe_download(url: str, output: str = '/content/new_file.txt', retries: int = 3, delay: int = 3, quiet: bool = False):
  for attempt in range(1, retries + 1):
      try:
          print(f"Intento #{attempt} de descarga...")
          gdown.download(url, output, quiet=quiet)
          print("Descarga exitosa")
          return True

      except PermissionError:
          print("Error: No tienes permiso para acceder al archivo")
          break  #No need to retry

      except requests.exceptions.RequestException as e:
          print(f"Error de Red: {e}")
          if attempt < retries:
              print(f"Reintentando en {delay} segundos...")
              time.sleep(delay)
          else:
              print("Eror de red: Máximo numéro de intentos alcanzados.")

      except OSError as e:
          print(f"Error al guardar el archivo localmente: {e}")
          break  # Local file system problem no retry
      except Exception as e:
          print(f"Error inesperado {e}")
          break

  return False  # If it exits the loop without returning True

#Download all needed information sources
#Optional if files are already included
def download_files():
  #Download MJL files if needed
  if not os.path.exists('/content/SCIE.csv'):
    safe_download("https://drive.google.com/uc?id=fileID", '/content/SCIE.csv', quiet=False)
  else:
    print("Archivo SCIE ya hallado")

  if not os.path.exists('/content/ESCI.csv'):
    safe_download("https://drive.google.com/uc?id=fileID", '/content/ESCI.csv', quiet=False)
  else:
    print("Archivo ESCI ya hallado")

  if not os.path.exists('/content/AHCI.csv'):
    safe_download("https://drive.google.com/uc?id=fileID", '/content/AHCI.csv', quiet=False)
  else:
    print("Archivo AHCI ya hallado")

  if not os.path.exists('/content/SSCI.csv'):
    safe_download("https://drive.google.com/uc?id=fileID", '/content/SSCI.csv', quiet=False)
  else:
    print("Archivo SSCI ya hallado")

  # Cuadro Maestro Excel file from google drive
  if not os.path.exists('/content/CuadroMaestro.xlsx'):
      safe_download("https://drive.google.com/uc?id=fileID", '/content/CuadroMaestro.xlsx', quiet=False)
  else:
    print("Archivo Cuadro Maestro ya hallado")

  #Inner database to adjoin group data
  if not os.path.exists('/content/Boletin.xlsx'):
      safe_download("https://drive.google.com/uc?id=fileID", '/content/Boletin.xlsx', quiet=False)
  else:
    print("Archivo Boletín ya hallado")


  #Global variables
  global mjl_df, publindex_found, boletin_df, cdrom_df
  scie_df = pd.read_csv("/content/SCIE.csv")
  esci_df = pd.read_csv("/content/ESCI.csv")
  ahci_df = pd.read_csv("/content/AHCI.csv")
  scii_df = pd.read_csv("/content/SSCI.csv")
  mjl_df = pd.concat([scie_df, esci_df, ahci_df, scii_df], ignore_index=True)
  boletin_df = pd.read_excel('/content/Boletin.xlsx',
                            sheet_name = 'INFORMACIÓN DE CARACTERIZACIÓN')
  cdrom_df = pd.read_excel('/content/CuadroMaestro.xlsx',
                            sheet_name = 'Investigacion - grupos y profe',
                           skiprows=4)

  #Default dictionary to save Publindex known results
  publindex_found = defaultdict(list)

#AUXILIARY FUNCTION

#Call on the extraction and verification of ISSN and creates the articles validation datafrmae
def search_indexed_journals(soup, code):
  results = []
  issn_list, articles = issn_list_from_gruplac(soup)
  print(f"issn found: {issn_list}")
  df_results = pd.DataFrame(articles, columns=["Artículos Publicados"])
  df_results[['Publindex', 'Scopus', 'WOS', 'Internacional']] = ''
  if issn_list:
    print("Verificando indexación...")
    for issn in issn_list:
      appears_publindex = check_publindex_indexing(issn)
      appears_international = ''
      appears_scopus = check_scopus_indexing(issn)
      appears_wos = check_mjl_indexed(issn)
      if appears_scopus or appears_wos:
        appears_international = 'X'
      results.append([appears_publindex, appears_scopus, appears_wos, appears_international])
      time.sleep(0.1)  # Avoid API rate limits

    print("Verificación finalizada")
    print(results)
    columns = ['Publindex', 'Scopus', 'WOS', 'Internacional']
    df_results[columns] = results
  #print("Resultados hallados: ")
  #display(df_results)

  total_row = pd.DataFrame([{
    'Artículos Publicados': 'Total',
    'Publindex': ((df_results['Publindex'] == 'X') & (df_results['Internacional'] == '')).sum(),
    'Scopus': (df_results['Scopus'] == 'X').sum(),
    'WOS': (df_results['WOS'] == 'X').sum(),
    'Internacional': (df_results['Internacional'] == 'X').sum()
    }])

  df_results = pd.concat([df_results, total_row], ignore_index=True)
  df_results.index = df_results.index + 1

  write_results(df_results, code, "1Oc4ybTMWO-Z5YXMe3VFHtW3nknqy81Na")

  return df_results, total_row


#Connection to grupLAC
def connect_gruplac(url):
  print(f"URL: {url}")
  found = False
  time.sleep(random.uniform(0.5, 2.0))
  headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/122.0.0.0 Safari/537.36"
  }

  if pd.isna(url):
    print("URL no válida")
    return [], []

  while not found:
    try:
      page = requests.get(url, headers=headers, timeout =30)
      found = True
      if page.status_code == 403:
        print(f"403 Forbidden for {url}")
        print("Retrying in 5 seconds")
        found = False
        time.sleep(5)

      if page.status_code != 200:
        print(f"Failed to load page: {url}  \nError code", page.status_code)
        found = False
        print("Retrying in 5 seconds")
        time.sleep(5)

    except RequestException as e:
      print(f"Error loading page: {url}")
      print(e)
      found = False
      print("Retrying in 5 seconds")
      time.sleep(5)

  print(f"Conexión exitosa!")

  soup = BeautifulSoup(page.text, "html.parser")
  #display(soup.text)

  return soup

#Extract ISSN list from the Gruplac url given the starting and ending year
def issn_list_from_gruplac(soup, start_year = 2020, end_year = 2024):

  #Array to collect resulting articles
  articles = []
  issn_list = []
  tables = soup.find_all("table")
  articles_table = None

  #Goes through all the tables finding checking the publication year
  for table in tables:
    header_cell = table.find("td", class_="celdaEncabezado")
    if header_cell and "Artículos publicados" in header_cell.text:
      articles_table = table
      break

  for row in articles_table.find_all("tr"):
    if len(row.find_all("td")) >= 2:
      content_cell = row.find_all("td")[1]
      text = content_cell.get_text(separator = " ", strip=True)
      year_match = re.search(r'ISSN:\s+(?:\d{4}-\d{3}[\dxX]|0),?\s+((19|20)\d{2})\s+', text)
      if year_match:
        #print("\nYear extracted: " + str(year_match.group(1)) + " " + text)
        art_year = int(year_match.group(1))
        if start_year <= art_year <= end_year:
          # Extract ISSN
          issn_match = re.search(r'ISSN:\s*(\d{4}-\d{3}[0-9X])', text, re.IGNORECASE)
          if issn_match:
            issn = issn_match.group(1)
            issn_list.append(issn)
            articles.append(text)

    #Print results for debugging
  #i = 1
  #print(f"\nArtículos entre {start_year} y {end_year}:")
  #for article in articles:
    #print(f"{i}. {article}")
    #i += 1

  return issn_list, articles

#Acceses the Publindex web page and looks up the existance of the ISSN obtained from the list
def check_publindex_indexing(issn):
  url = "https://scienti.minciencias.gov.co/publindex/api/publico/revistasPublindex/"
  payload = {
      "tipoBusqueda": "R",
      "txtIssn": issn,
  }

  headers = {
      "Content-Type": "application/json",
      "Accept": "application/json"
  }

  response = requests.post(url, json=payload, headers=headers, timeout=30)
  if response.status_code == 200:
    try:
      data = response.json()[0]
      if "ID_REVISTA" in data:
        return 'X'
    except RequestException as e:
      print(f"Error de lectura Publindex: {e}")
      return ''
  else:
    return ''


#Checks the existance of a given ISSN in the scopus database via the API
def check_scopus_indexing(issn):
    url = f"https://api.elsevier.com/content/serial/title?issn={issn}"
    headers = {
        "Accept": "application/json",
    }
    params =  {
      "apiKey" : "yourAPIKey",
      "insttoken": "yourInstToken",
    }
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        if "serial-metadata-response" in data:
          if 'error' in data['serial-metadata-response']:
            return ''
          elif 'entry' in data['serial-metadata-response']:
            return 'X'
    else:
      print(f"Error de lectura Scopus: {response.status_code}")

    return ''

#Compares the issn with the MJL csv files
def check_mjl_indexed(issn):
  global mjl_df
  if(issn in mjl_df['ISSN'].values) or (issn in mjl_df['eISSN'].values):
    return 'X'
  else:
    return ''

def count_basic_prod(soup, target_header="" , start_year=2020, end_year=2025):
  print(f"Counting {target_header}")

  tables = soup.find_all("table")
  target_table = None
  results = []

  #Goes through all the tables finding checking the publication year
  for table in tables:
    header_cell = table.find("td", class_="celdaEncabezado")
    if header_cell and header_cell.text.strip().lower() in target_header.lower():
      target_table = table
      break

  for row in target_table.find_all("tr"):
    if len(row.find_all("td")) >= 2:
      content_cell = row.find_all("td")[1]
      text = content_cell.get_text(separator = " ", strip=True)
      year_match = re.search(r',\s+((19|20)\d{2}),\s+', text)
      if year_match:
        #print("\nYear extracted: " + str(year_match.group(1)) + " " + text)
        art_year = int(year_match.group(1))
        if start_year <= art_year <= end_year:
          results.append(text)

  return len(results)

#Artistic products
def count_art_products(soup, start_year=2020, end_year=2024):
  print("Counting art products...")

  target_header = "Producción en arte, arquitectura y diseño"
  tables = soup.find_all("table")
  target_table = None
  results = { 0: [], 1: [], 2: [] }
  sections = {"Obras o productos":0, "Eventos Artísticos":1, "Talleres de Creación":2}

  for table in tables:
    header_cell = table.find("td", class_="celdaEncabezado")
    if header_cell and header_cell.text in target_header:
      target_table = table
      #print("Tabla encontrada: " + header_cell.text)
      break

  for row in target_table.find_all("tr"):
    header_cell = row.find("td", class_="celdaEncabezado")
    if header_cell:
      #print("Tabla: " + str(header_cell.text))
      section = sections.get(header_cell.text)

    if len(row.find_all("td")) >= 2:
      content_cell = row.find_all("td")[1]
      text = content_cell.get_text(separator = " ", strip=True)

      if section == 0 or section == 2:
        year_match = re.search(r'Fecha de creación:\s+\w+\s+de\s+((19|20)\d{2})\s+', text)
        #print("\nYear extracted: " + str(year_match.group(1)) + " " + text)
      elif section == 1:
        year_match = re.search(r'Fecha de inicio:\s+((19|20)\d{2})-', text)
      if year_match:
        #print("\nYear extracted: " + str(year_match.group(1)) + " " + text)
        art_year = int(year_match.group(1))
        if start_year <= art_year <= end_year:
          #print("Art product appended:" + text)
          results[section].append(text)

  print("Obras: " + str(len(results[0])))
  print("Eventos: " + str(len(results[1])))
  print("Talleres: " + str(len(results[2])))

  return len(results[0]), len(results[1]), len(results[2])

#Tutored works
def count_thesis(soup, start_year=2020, end_year=2024):
  print("Counting thesis...")

  target_header = "Trabajos dirigidos/turorías"
  tables = soup.find_all("table")
  target_table = None
  results = {"Trabajos de grado de pregrado": [],
              "Trabajo de grado de maestría o especialidad clínica" : [],
              "Tesis de doctorado": []}

  for table in tables:
    header_cell = table.find("td", class_="celdaEncabezado")
    if header_cell and header_cell.text in target_header:
      target_table = table
      break

  for row in target_table.find_all("tr"):
    if len(row.find_all("td")) >= 2:
      content_cell = row.find_all("td")[1]
      text = content_cell.get_text(separator = " ", strip=True)
      init_year_match = re.search(r'Desde\s+\d+\s+((19|20)\d{2})', text)
      final_year_match = re.search(r'hasta\s+\w+\s+((19|20)\d{2})', text)
      type_text = content_cell.find("strong")
      if type_text:
        type_text = type_text.text.strip()
      if init_year_match:
        #print("\nYear extracted: " + str(init_year_match.group(1)) + " " + text)
        art_year = int(init_year_match.group(1))
        if start_year <= art_year <= end_year:
          #print("Thesis appended:" + text)
          results.get(type_text, []).append(text)

  print("Pregrado: " + str(len(results["Trabajos de grado de pregrado"])))
  print("Maestría o especialidad clínica: " + str(len(results["Trabajo de grado de maestría o especialidad clínica"])))
  print("Doctorados: " + str(len(results["Tesis de doctorado"])))

  return len(results["Trabajos de grado de pregrado"]), len(results["Trabajo de grado de maestría o especialidad clínica"]), len(results["Tesis de doctorado"])

#Writes the results into a new sheet of the excel file
def write_results(df, code, file_id):
  file_path = '/content/CuadroMaestro.xlsx'

  # Create ExcelWriter with the loaded workbook
  with pd.ExcelWriter(file_path, mode='a', if_sheet_exists='replace', engine='openpyxl') as writer:
      df.to_excel(writer, sheet_name=code, index=False)

  media = MediaFileUpload(file_path, mimetype="application/vnd.openxmlformats-officedocument.spreadsheetml.sheet")
  updated_file = drive_service.files().update(fileId= file_id, media_body=media).execute()

#Writes the resulting rows
def write_row_by_row(df, file_id, number_rows):
  file_path = '/content/CuadroMaestro.xlsx'
  columns = [i for i in range(14) if i not in [0, 2, 3, 11]]

  wb = load_workbook(file_path)

  # Get the sheet safely
  if 'Investigacion - grupos y profe' not in wb.sheetnames:
      raise KeyError("Sheet 'Investigacion - grupos y profe' not found in workbook")

  ws = wb['Investigacion - grupos y profe']

  # Write data
  for row in range(5, 5 + number_rows):
      for col in columns:
          ws.cell(row=row+1, column=col+1, value = df.iloc[row-5, col])

  wb.save(file_path)
  media = MediaFileUpload(file_path, mimetype="application/vnd.openxmlformats-officedocument.spreadsheetml.sheet")
  updated_file = drive_service.files().update(fileId= file_id, media_body=media).execute()
```
